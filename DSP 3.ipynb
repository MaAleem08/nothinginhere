OOP STYLE METHOD

```python
import numpy as np
x = np.linspace(0, 2, 100) # no. of samples to generate

#Step 1: Declaring the figure and the axis
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.

#Step 2: start feeding data
ax.plot(x, x, label='linear')  # Plot some data on the axes. #plot 1
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes... #plot 2
ax.plot(x, x**3, label='cubic')  # ... and some more.

#step 3: Naming your x and y axis
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.

ax.set_title("Simple Plot")  # Add a title to the axes.

ax.legend()  # Add a legend.
```

PYPLOT STYLE

```python
x = np.linspace(0, 2, 100)#Return evenly spaced numbers over a specified interval

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()
```

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot 
plt.plot(x,y, marker="o", color="r", linestyle="None", ms = 20, mfc = 'y', mec = 'r') 
  
# function to show the plot 
plt.show() 
```

```python
# x values 
x = [5, 2, 9, 4, 7] 
  
# Y values 
y = [10, 5, 8, 4, 2] 
  
# Z values
z = [15, 8, 7, 5, 6] 

fig ,ax = plt.subplots()
# Function to plot 
ax.plot(x, y, marker="v", color="r", label="first plot")#st line plot

ax.plot(y, z, marker="p", color="b", linestyle="-.", label="second plot") #2nd line plot

ax.plot(x, z, marker="+", color="g", label="th ird plot") #3rd line plot


ax.set_xlabel('Height')
ax.set_ylabel('age')

plt.title("Simple Plot")
ax.legend()
  
# function to show the plot 
plt.show() 
```

```python
x = [5, 2, 9, 4, 7] 
  
# Y values 
y = [10, 5, 8, 4, 2] 
  
# Z values
z = [15, 8, 7, 5, 6] 


fig, ax = plt.subplots(3, 1)

ax[0].plot(x, y, marker="v", color="r", linestyle="--")

ax[1].plot(y, z, marker="o", color="b", linestyle="-.")

ax[2].plot(x, z, marker="+", color="g", linestyle="-")

ax[0].set_ylabel("xlabel")
ax[1].set_ylabel("ylabel")
ax[2].set_ylabel("z label")

ax[1].set_xlabel("Simple Plot")
ax[2].set_xlabel("Simple Plot")

# function to show the plot
plt.show()
```
