```python
import re

def palindrome(s):
    if type(s) == str :
    
        s = re.sub(r'[^a-zA-Z0-9]', '', s.lower())

        return s == s[::-1]

    else:
        print("give a string")
        
input_string =  "A man, a plan, a canal: Panama"
print(palindrome(input_string))
```

## question 3

```python
import pandas as pd , numpy as np
from scipy.stats import ttest_1samp
```

```python
df = pd.read_csv("FuelConsumption.csv")
df.head()
```

```python
# lets say someone is saying that the avg fuel consumpotion is not equal 14.
# we want to know if the fuel consumotion is different from that

# h0 : avg fuel consumpion = 14
# h1 : avg fuel consumpion != 14
# alpha 0.025 bcoz its two sided

t_statistic, p_value = ttest_1samp(df["FUEL CONSUMPTION"] , 14 , alternative='two-sided')
print(p_value)

# p value is way less than alpha so we reject null hypothesis
# avg fuel consumption != 14
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df.drop(columns=["Year","MODEL"],inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(columns=["MAKE","VEHICLE CLASS","TRANSMISSION","FUEL"],data = df , drop_first=True)
df.head()
```

```python
ss = StandardScaler()
df[["ENGINE SIZE","COEMISSIONS "]] = ss.fit_transform(df[["ENGINE SIZE","COEMISSIONS "]])
df.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=["FUEL CONSUMPTION"]) ,df["FUEL CONSUMPTION"],test_size=0.25,random_state=123)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
print(np.sqrt(mse(y_test , pred)))
```

```python
# the root mean squared error is 0.067
```
